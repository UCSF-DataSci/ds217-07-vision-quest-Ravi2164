### Part 1.1: Basic Figures and Subplots

**TODO: Create a figure with 2x2 subplots showing different plot types**

```python
# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x/5)
y4 = np.log(x + 1)

# TODO: Create a 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# TODO: Plot on each subplot
# TODO: Add grid to each subplot
# Top-left: Sine wave
axes[0, 0].plot(x, y1, label='sin(x)', color='red', linewidth=2)
# Top-right: Cosine wave 
axes[0, 1].plot(x, y2, label='cos(x)', color='blue', linewidth=2, linestyle='--') 
# Bottom-left: Exponential decay
axes[1, 0].plot(x, y3, label='exp(-x/5)', color='green', linewidth=2)
# Bottom-right: Logarithmic function
axes[1,1].plot(x, y4, label='log(x + 1)', color='green', linewidth=2)
# TODO: Add titles and labels to each subplot
#top-left
axes [0, 0].plot(x, y1)
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X values')
axes[0, 0].set_ylabel('Sin(x)')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].legend()
#top-right
axes [0, 1].plot(x, y2)
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('X values')
axes[0, 1].set_ylabel('cos(x)')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].legend()
#Bottom-left
axes [1, 0].plot(x, y3)
axes[1, 0].set_title('Exponential decay')
axes[1, 0].set_xlabel('X values')
axes[1, 0].set_ylabel('exp(-x/5)')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].legend()
#bottom-right
axes [1, 1].plot(x, y4)
axes[1, 1].set_title('Logarithmic function')
axes[1, 1].set_xlabel('X values')
axes[1, 1].set_ylabel('log(x+1)')
axes[1, 1].grid(True)
axes[1, 1].legend()
# TODO: Use tight_layout() to prevent overlapping
plt.tight_layout()
plt.show()
```

## Question 2: seaborn Statistical Visualization

### Part 2.1: Load and Explore Data

```python
# Load the sales data
sales_df = pd.read_csv('data/sales_data.csv')
customer_df = pd.read_csv('data/customer_data.csv')
product_df = pd.read_csv('data/product_data.csv')

# Merge data for analysis
merged_df = sales_df.merge(customer_df, on='customer_id').merge(product_df, on='product_id')

print("Data shape:", merged_df.shape)
print("\nColumns:", merged_df.columns.tolist())
print("\nFirst few rows:")
print(merged_df.head())
```

### Part 2.2: Statistical Plots

**TODO: Create statistical visualizations with seaborn**

```python
# TODO: Create a figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# TODO: Top-left: Box plot of total_amount by store_location
sns.boxplot(data=merged_df, x='store_location', y='total_amount', ax=axes[0, 0], color='lightblue' \
'')
axes[0, 0].set_title('Amount by Store')
# TODO: Top-right: Violin plot of total_amount by category
sns.violinplot(data=merged_df, x='store_location', y='total_amount', ax=axes[0, 1], color='green')
axes[0, 1].set_title('Amount by Store')
# TODO: Bottom-left: Scatter plot of quantity  vs total_amount with hue by gender
sns.scatterplot(data=merged_df, x='quantity', y='total_amount', 
                hue='gender', ax=axes[1, 0], color='red')
axes[1, 0].set_title('Quantity vs Total Amount')

# TODO: Bottom-right: Histogram of total_amount with kde overlay
sns.histplot(data=merged_df, x='total_amount', kde=True, color='pink',
             alpha=0.7, ax=axes[1, 1])
axes[1, 1].set_title('Total Amount Distribution')
# TODO: Add appropriate titles and labels, 
# TODO: Customize colors and styling
# TODO: Save the plot as 'output/q2_seaborn_plots.png'

plt.tight_layout()
plt.savefig('output/q2_seaborn_plots.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Part 2.3: Correlation Analysis

**TODO: Create a correlation heatmap**

```python
# TODO: Select numeric columns for correlation analysis
# First, explore the merged data to see what columns are available
print("Available columns:", merged_df.columns.tolist())
print("Data types:")
print(merged_df.dtypes)

# TODO: Select only the numeric columns that exist for correlation analysis:
# ('quantity','unit_price_x','total_amount','age', 'unit_price_y') 
# Hint: After merging, pandas adds _x and _y suffixes to duplicate column names

numeric_cols = ['quantity','unit_price_x','total_amount','age', 'unit_price_y']  # Fill this with the appropriate column names
numeric_data = merged_df[numeric_cols]
# TODO: Calculate correlation matrix
# correlation_matrix = ...
correlation_matrix = numeric_data.corr()
print (correlation_matrix) 

# TODO: Create heatmap with seaborn
# TODO: Add title and customize appearance
plt.figure(figsize=(6, 4))
sns.heatmap(data=correlation_matrix, annot= True, cmap= 'coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.xlabel("Columns")
plt.ylabel("columns")

# TODO: Save the plot as 'output/q2_correlation_heatmap.png'
plt.tight_layout()
plt.savefig('output/q2_correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Question 3: pandas Plotting and Data Exploration

### Part 3.1: Time Series Visualization

**TODO: Create time series plots with pandas**

```python
# TODO: Convert transaction_date to datetime
# merged_df['transaction_date'] = pd.to_datetime(merged_df['transaction_date'])
merged_df['transaction_date'] = pd.to_datetime(merged_df['transaction_date'])
#print(merged_df.dtypes)     
                                       
# TODO: Create daily sales aggregation
# daily_sales = merged_df.groupby('transaction_date')['total_amount'].sum()
daily_sales = merged_df.groupby('transaction_date')['total_amount'].sum()
# TODO: Create a figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# TODO: Top-left: Line plot of daily sales
daily_sales.plot(ax=axes[0, 0], color='red', linestyle='-')
axes[0, 0].set_title("Daily Sales")
axes[0, 0].set_ylabel("Total Amount")
axes[0, 0].set_xlabel("Date")
axes[0, 0].grid(True)
# TODO: Top-right: Rolling 7-day average
# average:
daily_sales_7day = daily_sales.rolling(window=7).mean()
daily_sales_7day.plot(ax=axes[0, 1] ,color='blue', linestyle='-')
axes[0, 1].set_title("7 -Day Average of Daily Sales")
axes[0, 1].set_ylabel("Total Amount")
axes[0, 1].set_xlabel("Date")
axes[0, 1].grid(True)

# TODO: Bottom-left: Monthly sales bar chart (use resample('ME') for month-end frequency)
#monthly sales
monthly_sales= daily_sales.resample('M').sum()

monthly_sales.plot(ax=axes[1, 0] ,color='green', linestyle='-', kind='bar')
axes[1, 0].set_title("Monthly Sales")
axes[1, 0].set_ylabel("Total Amount")
axes[1, 0].set_xlabel("Months")
axes[1, 0].grid(True)

# TODO: Bottom-right: Sales distribution histogram
daily_sales.plot(ax=axes[1, 1] ,color='pink', bins=20, kind='hist')

axes[1, 1].set_title("Sales Distribution")
axes[1, 1].set_ylabel("Total Amount")
axes[1, 1].set_xlabel("Frequency")
axes[1, 1].grid(True)
# TODO: Add titles and labels
# TODO: Save the plot as 'output/q3_pandas_plots.png'

plt.tight_layout()
plt.savefig('output/q3_pandas_plots.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Part 3.2: Data Overview Dashboard

**TODO: Create a comprehensive data overview**

```python
# TODO: Create a figure with multiple subplots
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes= axes.flatten()

# TODO: Plot 1: Sales by category (bar chart)
sales_by_category = merged_df.groupby('category') ['total_amount']. sum()
sales_by_category.plot(kind='bar', color='blue', ax= axes[0])
axes[0].set_title("Sales by Category")
axes[0].set_ylabel("Total Amount")
axes[0].set_xlabel("Category")
axes[0].grid(True)
# TODO: Plot 2: Sales by store location (pie chart)
sales_by_store = merged_df.groupby('store_location') ['total_amount']. sum()
sales_by_store.plot(kind='pie', ax= axes[1], autopct='%1.1f%%')
axes[1].set_title("Sales by Store Location")
axes[1].set_ylabel("Total Amount")
axes[1].set_xlabel("Store Location")
axes[1].grid(True)
# TODO: Plot 3: Age distribution (histogram)
merged_df['age'].plot(ax=axes[2] ,color='pink', bins=20, kind='hist') 
axes[2].set_title("Age Distribution")
axes[2].set_ylabel("Count")
axes[2].set_xlabel("Age")
axes[2].grid(True)
# TODO: Plot 4: Quantity vs total_amount scatter
merged_df.plot(kind= 'scatter',x='quantity', y='total_amount', color='orange', ax=axes [3])
axes[3].set_title("Quantity vs Total Amount")
axes[3].set_ylabel("Quantity")
axes[3].set_xlabel("Total Amount")
axes[3].grid(True)

# TODO: Plot 5: Sales over time (line plot)
daily_sales.plot(kind='line', ax=axes[4], color='red', linestyle='-')
axes[4].set_title("Sales over time")
axes[4].set_ylabel("Total Amount")
axes[4].set_xlabel("Date")
axes[4].grid(True)

# TODO: Plot 6: Top 10 products by sales (horizontal bar)
top_products = merged_df.groupby('product_name')['total_amount'].sum().sort_values(ascending=True).tail(10)
top_products.plot(kind='barh', color='green', ax=axes[5])
axes[5].set_title("Top 10 products by Sales")
axes[5].set_ylabel("Products")
axes[5].set_xlabel("total sales")
axes[5].grid(True)

# TODO: Add titles and customize appearance
# TODO: Save the plot as 'output/q3_data_overview.png'

plt.tight_layout()
plt.savefig('output/q3_data_overview.png', dpi=300, bbox_inches='tight')
plt.show()
```
